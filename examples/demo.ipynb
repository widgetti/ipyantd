# Widget interface


```python
import ipywidgets
from IPython.display import clear_output

options = [
    dict(label="Apple", value="apple"),
    dict(label="Pear", value="pear"),
    dict(label="Cherry", value="cherry"),
]

menu_items = [
    dict(key="1", label=ipyreact.Widget(_type="div", children=["Click me 1"], events={"onClick": print})),
    dict(key="2", label=ipyreact.Widget(_type="div", children=["Click me 2"], events={"onClick": print})),
]
text = """Lorem ipsum dolor sit amet, consectetur adipiscing elit.
Sed non risus. Suspendisse lectus tortor, dignissim sit amet, adipiscing nec, ultricies sed, dolor. 
Cras elementum ultrices diam. Maecenas ligula massa, varius a, semper congue, euismod non, mi. 
Proin porttitor, orci nec nonummy molestie, enim est eleifend mi, non fermentum diam nisl sit amet erat. 
Duis semper.
Duis arcu massa, scelerisque vitae, consequat in, pretium a, enim. Pellentesque congue.
"""


def dialog_open(event_data):
    with output:
        clear_output(wait=True)
        print("Modal opened")
    dialog.value = True


def dialog_close(event_data):
    with output:
        clear_output(wait=True)
        print("Modal closed")
    dialog.value = False
    
def show_value(change):
    new = change["new"]
    with output:
        clear_output(wait=True)
        print("value = ", new)

def report_click(event_data=None):
    with output:
        clear_output(wait=True)
        print("Clicked")

demo = ipyreact.Widget(_type="div", children=[
    antd.Row(props=dict(gutter=[48,48], style={"padding": "100px"}), children=[
        antd.Col(props=dict(span=8, style={"boxSizing": "border-box"}), children=[
            ipyreact.Widget(_type="div", children=[
                antd.Button(children=["Button"], props=dict(type="primary"), events={"onClick": report_click}),
                antd.Button(children=["Button"]),
            ]),
        ]),
        antd.Col(props=dict(span=8), children=[
            antd.Button(children=["Button"], props=dict(type="primary", shape="round")),
            antd.Button(props=dict(type="primary", shape="circle", icon=icons.Icon("SearchOutlined"))),
            antd.Button(props=dict(shape="circle", icon=icons.Icon("SearchOutlined"))),
        ]),
        antd.Col(props=dict(span=8), children=[
            antd.Button(children=["Button"], props=dict(type="text")),
        ]),
        antd.Col(props=dict(span=8), children=[
            antd.Button(children=["Button"], props=dict(type="primary", size="small")),
            antd.Button(children=["Button"], props=dict(size="small")),
        ]),
        antd.Col(props=dict(span=8), children=[
            antd.Button(children=["Button"], props=dict(type="primary", shape="round", size="small")),
            antd.Button(props=dict(type="primary", shape="circle", icon=icons.Icon("SearchOutlined"), size="small")),
            antd.Button(props=dict(shape="circle", icon=icons.Icon("SearchOutlined"), size="small")),
        ]),
        antd.Col(props=dict(span=8), children=[
            antd.Button(children=["Button"], props=dict(type="text", size="small")),
        ]),
        
        # sliders
        antd.Col(props=dict(span=8), children=[
            antd.Slider(value=0, props=dict(min=0, max=100)),
        ]),
        antd.Col(props=dict(span=8), children=[
            slider_range := antd.Slider(value=[20, 70], props=dict(min=0, max=100, range=True)),
        ]),
        antd.Col(props=dict(span=8), children=[
            slider := antd.Slider(value=30, props=dict(min=0, max=100, tooltip=dict(open=True))),
        ]),

        # select
        antd.Col(props=dict(span=8), children=[
            select := antd.Select(value="pear", props=dict(options=options, style=dict(width="120px"))),
        ]),
        antd.Col(props=dict(span=8), children=[
            multi_select := antd.Select(value=["cherry"], props=dict(options=options, style=dict(width="220px"), mode="multiple"))
        ]),
        antd.Col(props=dict(span=8), children=[
        # only supported in 5.31+
        #   antd.Select(value="pear", props=dict(options=options, variant="filled", style=dict(width="120px"))),
        ]),
        
        # misc
        antd.Col(props=dict(span=8), children=[
            switch := antd.Switch(value=True),
        ]),
        antd.Col(props=dict(span=8), children=[
            dropdown := antd.Dropdown(props=dict(menu=dict(items=menu_items), trigger=["click"]), children=
                [ipyreact.Widget(_type="span", children=["Menu"])]
            ),
        ]),
        antd.Col(props=dict(span=8), children=[
            dialog := antd.Modal(events={"onOk": dialog_close, "onCancel": dialog_close}, children=[
                ipyreact.Widget(_type="div", children=[text])
            ]),
            antd.Button(children=["Dialog"], props=dict(type="primary"), events={"onClick": dialog_open}),
        ]),

        
    ]),
    output := ipywidgets.Output()
])
select.observe(show_value, "value")
multi_select.observe(show_value, "value")
slider.observe(show_value, "value")
slider_range.observe(show_value, "value")
switch.observe(show_value, "value")
with output:
    print(" ")
demo
```
